## **Import Libraries**

```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## **Load Dataset**

```python
# Load Iris dataset from seaborn
iris = sns.load_dataset("iris")

# Display shape of dataset
print("Dataset Shape:", iris.shape)

# Display column names
print("Columns:", iris.columns)
```

## **Preview Dataset**

```python
# Display first five rows
iris.head()
```

## **Dataset Information**

```python
# Display dataset information
iris.info()
```

## **Summary Statistics**

```python
# Generate summary statistics
iris.describe()
```

## **Scatter Plot**

```python
# Scatter plot showing relationship between sepal length and petal length
plt.figure()

sns.scatterplot(
    x="sepal_length",
    y="petal_length",
    hue="species",
    data=iris
)

plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")

plt.show()
```

## **Histograms**

```python
# Histogram to show distribution of features
iris.hist(figsize=(10,8))

plt.suptitle("Feature Distributions in Iris Dataset")

plt.show()
```

## **Box Plot (Outliers Detection)**

```python
# Box plot to detect outliers
plt.figure(figsize=(8,6))

sns.boxplot(data=iris)

plt.title("Box Plot of Iris Features")

plt.show()
```

## **Conclusion**

### Insights:

### 1. Petal length and petal width are strongly correlated.
### 2. Setosa species has smaller petals compared to other species.
### 3. Box plots indicate a few possible outliers in petal measurements.
### 4. Visualizations help understand how different species vary in size.
